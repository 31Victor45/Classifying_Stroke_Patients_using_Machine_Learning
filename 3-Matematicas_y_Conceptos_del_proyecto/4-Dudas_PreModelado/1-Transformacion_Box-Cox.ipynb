# Transformación Box-Cox: Domando la Variabilidad y el Sesgo

Muchos modelos de Machine Learning (como la Regresión Lineal o el LDA) asumen que los datos siguen una **Distribución Normal** (campana de Gauss) y que tienen **Homocedasticidad** (varianza constante). 

Sin embargo, en el mundo real, los datos suelen estar "sesgados" o tener varianza que crece con el tiempo. Aquí es donde entra **Box-Cox**.

## 1. ¿Qué es la Transformación Box-Cox?

Es una transformación matemática definida por un parámetro llamado **Lambda ($\lambda$)**. Su objetivo es transformar datos no normales en una distribución lo más normal posible.

La fórmula general es:

$$y(\lambda) = \begin{cases} \frac{y^\lambda - 1}{\lambda} & \text{si } \lambda \neq 0 \\ \ln(y) & \text{si } \lambda = 0 \end{cases}$$

> **Nota Crítica:** Box-Cox solo funciona con datos **estrictamente positivos** ($y > 0$). Si tienes valores de cero o negativos, debes sumar una constante antes de aplicarla.

---

## 2. ¿Por qué es útil en Machine Learning?

1. **Normalización de Residuos:** Los modelos lineales funcionan mejor si la relación entre variables no es "explosiva" (exponencial).
2. **Estabilización de la Varianza:** Evita que el modelo le dé demasiada importancia a los valores más grandes solo porque tienen una dispersión mayor.
3. **Mejora del R-cuadrado:** Al hacer la relación más lineal, el modelo puede capturar mejor la tendencia.
4. **Detección de Outliers:** Al comprimir las colas de la distribución, es más fácil identificar qué puntos son realmente extraños.
