# SMOTE + Limpieza: Elevando el nivel con Tomek Links y ENN

El problema del **SMOTE** básico es que crea datos nuevos de forma algo "ciega". A veces, genera datos sintéticos en lugares donde se mezclan demasiado con la clase mayoritaria, creando **ruido** y confusión para el modelo.

Para solucionar esto, surgieron las técnicas **híbridas**: primero sobremuestreamos (SMOTE) y luego limpiamos la "frontera" entre las clases.

---

## 1. SMOTE + Tomek Links: La Limpieza de Frontera

Para entender esta técnica, primero debemos saber qué es un **Tomek Link**.

### ¿Qué es un Tomek Link?
Se dice que existe un "vínculo de Tomek" cuando dos ejemplos de clases distintas son los **vecinos más cercanos** el uno del otro. Es como si dos enemigos estuvieran viviendo pared de por medio; esa zona es muy ambigua.

### ¿Cómo funciona el híbrido?
1. **Fase SMOTE:** Crea datos sintéticos para aumentar la clase minoritaria.
2. **Fase Tomek:** Identifica todos los "Tomek Links" (parejas de clases distintas muy pegadas) y **elimina el ejemplo de la clase mayoritaria** (o ambos).

**Resultado:** Se crea un "espacio en blanco" o zona de seguridad entre las dos clases. El modelo ahora puede dibujar una línea de separación mucho más clara.

## 2. SMOTE + ENN (Edited Nearest Neighbors)

Esta técnica es un poco más agresiva y profunda que Tomek Links.

### ¿Qué es ENN?
El método de **Vecinos Cercanos Editados** observa a un grupo de vecinos (usualmente 3). Si un ejemplo está rodeado de vecinos que pertenecen a una clase distinta, se considera ruido y se elimina.

### ¿Cómo funciona el híbrido?
1. **Fase SMOTE:** Genera los datos sintéticos para equilibrar las clases.
2. **Fase ENN:** Revisa todo el dataset. Si un dato (ya sea original o sintético) tiene vecinos que mayoritariamente pertenecen a la otra clase, **se borra**.

**Diferencia clave:** Mientras que Tomek solo limpia la "línea divisoria", ENN limpia **áreas enteras** de confusión. Es una limpieza mucho más exhaustiva.

## 3. Reflexión Didáctica: ¿Por qué son mejores que el SMOTE simple?

Imagina que estás enseñando a un niño a distinguir entre manzanas rojas y manzanas verdes:

* **SMOTE:** Simplemente pone más manzanas verdes en la mesa, pero algunas quedan mezcladas con las rojas. El niño se confunde.
* **SMOTE + Tomek:** Pone más manzanas verdes y quita las rojas que están tocando a las verdes. El niño ve una separación clara.
* **SMOTE + ENN:** Pone más manzanas verdes y quita cualquier manzana (de cualquier color) que esté en el lugar equivocado. El niño ve dos grupos perfectamente separados.

---

### Nota Importante sobre Validación
Cuando uses estas técnicas, **solo aplícalas al conjunto de ENTRENAMIENTO (Train)**. Nunca toques el conjunto de prueba (Test), ya que el Test debe representar la realidad (con su ruido y su desbalanceo) para que tu evaluación sea honesta.
