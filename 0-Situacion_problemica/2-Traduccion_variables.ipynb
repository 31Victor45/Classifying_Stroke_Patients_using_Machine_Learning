## 2. Diccionario de Datos (Atributos Clínicos)

Para el desarrollo del modelo, el departamento de sistemas del hospital ha proporcionado el siguiente desglose de variables recopiladas durante la admisión de los pacientes:

| Variable | Descripción | Valores / Unidades |
| :--- | :--- | :--- |
| **id** | Identificador único | Numérico (Se eliminará durante el entrenamiento) |
| **gender** | Género del paciente | "Male" (Masculino), "Female" (Femenino), "Other" |
| **age** | Edad | Años cumplidos |
| **hypertension** | Historial de hipertensión | 0: No padece, 1: Padece hipertensión |
| **heart_disease** | Enfermedades cardíacas | 0: Sin historial, 1: Con historial cardíaco |
| **ever_married** | Estado civil previo | "No" o "Yes" (¿Se ha casado alguna vez?) |
| **work_type** | Tipo de ocupación | "children" (Menor de edad), "Govt_jov" (Empleado público), "Never_worked" (Nunca trabajó), "Private" (Sector privado), "Self-employed" (Independiente) |
| **Residence_type** | Entorno de vivienda | "Rural" o "Urban" (Urbano) |
| **avg_glucose_level** | Nivel de glucosa | Nivel promedio de glucosa en sangre ($mg/dL$) |
| **bmi** | IMC | Índice de Masa Corporal ($kg/m^2$) |
| **smoking_status** | Relación con el tabaco | "formerly smoked" (Exfumador), "never smoked" (Nunca fumó), "smokes" (Fumador actual) o "Unknown" (Desconocido)* |
| **stroke** | **Variable Objetivo** | 1: El paciente sufrió un infarto, 0: No sufrió infarto |

> **Nota sobre 'Unknown':** En la variable `smoking_status`, el valor "Unknown" indica que la información no estaba disponible en el momento del ingreso o el paciente no pudo proporcionarla. Deberá tratarse con cuidado durante la fase de limpieza.
